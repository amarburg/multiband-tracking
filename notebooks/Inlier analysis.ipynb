```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal_dir = "../data/single_band_1fps/"
whiteout_dir = "../data/single_band_whiteout/"

names = ["file","inliers"]
normal = pd.read_csv( normal_dir + "homography.csv", names=names)
whiteout = pd.read_csv( whiteout_dir + "homography.csv", names=names)

## Load and append detections
names = ["file","detections"]
n = pd.read_csv( normal_dir + "detections.csv", names=names)
normal = pd.concat( [n, normal["inliers"]], axis="columns" )

w = pd.read_csv( whiteout_dir + "detections.csv", names=names)
whiteout = pd.concat( [w, whiteout["inliers"]], axis="columns" )


normal["inlier_pct"] = normal["inliers"]/normal["detections"]
whiteout["inlier_pct"] = whiteout["inliers"]/whiteout["detections"]

print(normal)
print(whiteout)


```

```python
plt.figure(figsize=(12,4))

plt.plot( normal["detections"], label="original")
plt.plot( whiteout["detections"], label="whiteout")

plt.legend()
plt.title("Number of ORB features extracted")
plt.ylabel("Num features")
plt.xlabel("Image sequence num")

plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.plot( normal["inlier_pct"], label="original")
plt.plot( whiteout["inlier_pct"], label="whiteout")

plt.legend()
plt.title("Pct inliers used to calculate image alignment")
plt.ylabel("Pct features used")
plt.xlabel("Image sequence num")

plt.show()
```
